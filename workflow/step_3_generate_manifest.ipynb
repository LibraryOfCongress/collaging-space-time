# Step 3: Generate Manifest JSON

Aggregate the Metadata JSON and Model Results JSON into a manifest JSON, with all of the information about the digital assets.

* In the manifest, each row is an item including of all the metadata below and a list of segments generated by the computer vision models.

* Each row of the manifest should include:
    * Item Title
    * URL
    * subject List
    * date
    * alt_text entry -- duplicate the item title for now.
    * contributor_names
    * Locations 
    * Collection Name (source_collection)
        *  set -- specific to the Free-to-Reuse
    * original_format
    * thumbnail (filename)
    * List of Segements (Filename of the cutout, Filename of the mask, and bounding box information)
        *   outputs from step 1 and step 2 (filename and bounding box information)

```python
import json


def read_json(file_path):
    with open(file_path, "r") as f:
        data_dict = json.load(f)
    return data_dict
```

```python
metadata = read_json("workflow_data/items_metadata.json")
model_data = read_json("workflow_data/model_results.json")
```

```python
manifest = model_data.copy()

print(len(manifest["items"]))
print(len(metadata.keys()))
```

```python
for i in metadata:
    # print(item)
    for j in manifest["items"]:
        # print(j['resource_id'],metadata[i]['resource_id'])
        if metadata[i]["resource_id"] == j["resource_id"]:
            # print(metadata[i]["resource_id"])
            # print(metadata[i]["dates"])
            j.update(
                {
                    "resource_id": metadata[i]["resource_id"],
                    "title": metadata[i]["title"],
                    "item_url": f"https://loc.gov/item/{metadata[i]['resource_id']}/",
                    "image_url": metadata[i]["item_url"],
                    "subjects": metadata[i]["subjects"],
                    "dates": metadata[i]["dates"],
                    "contributors": metadata[i]["contributors"],
                    "description": metadata[i]["description"],
                    "alt_text": metadata[i]["alt_text"],
                    "collections": metadata[i]["collection"],
                }
            )
        else:
            continue
```

```python
with open(f"../ui/data/manifest.json", "w") as f:
    json.dump(manifest, f, indent=4)
print("Manifest JSON created")
```
